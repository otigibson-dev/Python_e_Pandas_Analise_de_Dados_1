```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_excel("/content/drive/MyDrive/datasets/AdventureWorks.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Qual foi a receita total
df["Valor Venda"].sum()
```

```python
df["Custo Final"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
#custo total:

round(df["Custo Final"].sum(), 2)
```

```python
#Lucro

df["Lucro"] = df["Valor Venda"] - df["Custo Final"]
```

```python
df.head()
```

```python
round(df["Lucro"].sum(), 2)
```

```python
#Quanto tempo levou para o produto chegar ao cliente:
df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

```python
#Quero saber, agora, a média de tempo de envio para cada marca
#Vou precisar transformar a coluna de tempo em numerica

df["Dias Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
#agora pra pegar a media

df.groupby("Marca")["Dias Envio"].mean().plot()
```

```python
df.isnull().sum()
```

Para saber o lucro por ano e por marca


```python
#primeiro preciso agrupar por ano e marca

df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().plot()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"]) ["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?

df.groupby("Produto") ["Quantidade"].sum().sort_values(ascending=False)
```

```python
#gafico Total de Produtos Vendidos:

df.groupby("Produto") ["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#gerando grafico do lucro por ano

df.groupby(df["Data Venda"].dt.year) ["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca") ["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
